# HNMR spectra per dietary group

This notebook reproduces figure 2 of the paper.

```python
from matplotlib import pyplot as plt

from smelly_rats.loader import load_smelly_rats
```

```python
hnmr_spectra, target = load_smelly_rats('../data/onion_NMR.mat')
```

```python
plotr = (
    hnmr_spectra
    .groupby(target, axis=1)
    .mean()
)

fig, axes = plt.subplots(plotr.shape[1], figsize=(10, 8))

for idx, ax in enumerate(axes):
    plotr.iloc[:, idx].plot(ax=ax)

    ax.set_xlim([10, 0])
    ax.set_ylim([0, 45_000])
    ax.set_title(f'Onion type {plotr.columns[idx]}')
    ax.set_xticks([])
    
axes[-1].set_xticks(list(range(9, 0, -1)))
    
fig.tight_layout()
```
